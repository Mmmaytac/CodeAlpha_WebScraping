## Importing libraries

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.service import Service
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
from datetime import date
import pandas as pd
import numpy as np
```

## Set Up Chrome WebDriver

```python
driver=webdriver.Chrome()
url="https://tap.az/elanlar/neqliyyat/avtomobiller"
driver.get(url)
```

```python
price=driver.find_element(By.CSS_SELECTOR, ".price-val" )
print(price.text)
```

## Data Extraction

```python
elements=driver.find_elements(By.CSS_SELECTOR, 'div.products-i')
elements
```

## Creating DataFrame

```python
lst=[element.text for element in elements]
```

```python
new_lst=[]
for row in lst:
    row.replace(",","\n")
    new_lst.append(row.split("\n"))
new_lst
```

```python
columns=["bookmarks","price","name/year","city/insert_date/time"]
df=pd.DataFrame(data=new_lst, columns=columns)
df
```

## Cleaning and Structuring Data

```python
df[["name","year"]]=df["name/year"].str.split(",",n=1,expand=True)
df[["city","insert_date","time"]]=df["city/insert_date/time"].str.split(",",expand=True)
df.drop(columns=["name/year","city/insert_date/time","bookmarks"],inplace=True)
df
```

```python
df.columns
```

```python
columns=[]
for i in df.columns:
    columns.append(i.capitalize())
df.columns=columns
```

```python
df
```

```python
df.dropna(inplace=True)
```

## Data Normalization

```python
df["Year"]=df["Year"].apply(lambda x: x[:5]).astype('int')
```

```python
df['Time'] = pd.to_datetime(df['Time'].str.strip(), format='%H:%M').dt.time
```

```python
df['Price']=df['Price'].str.replace(' ',"")
df['Price']=df['Price'].str.replace('AZN',"")
df['Price']=df['Price'].astype("int")
```

```python
df
```

```python
df.info()
```

```python
today=date.today()
df.Insert_date=df.Insert_date.str.replace("bugün",str(today))
```

```python
df.reset_index(drop=True,inplace=True)
df
```

## Exporting Data to CSV

```python
df.to_csv('tapaz_cars.csv', index=False)
print("Data has been exported to 'tapaz_cars.csv'")
```
